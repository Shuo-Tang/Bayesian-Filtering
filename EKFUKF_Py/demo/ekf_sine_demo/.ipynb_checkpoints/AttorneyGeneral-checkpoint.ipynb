```python
# Compute relative distance from the mean of other scores
pg_std = pg.copy()
for i in range(6):
    others = pg.iloc[list(set([0,1,2,3,4,5]) - {i})]
    pg_std.iloc[i] = (pg.iloc[i]-others.mean())/others.std()
```

```python
def color_negative_red(val):
    color = 'red' if np.abs(val) > 2 else 'black'
    return 'color: %s' % color
```

```python
def highlight_max(s):
    is_max = (np.abs(s) == np.abs(s).max()) & (np.abs(s) > 2)
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
pg_std.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```

```python
pg_std.mean().sort_values(ascending=False)
```
